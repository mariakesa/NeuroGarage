```python
import numpy as np
import matplotlib.pyplot as plt

# True parameters
true_mean = 5
true_std = 2

# Number of samples to simulate the process
max_sample_size = 100000
step_size = 10  # Increase the sample size by this amount in each step

# Arrays to hold sample sizes and their corresponding MLE estimates of the mean
sample_sizes = np.arange(step_size, max_sample_size + 1, step_size)
mle_estimates = []

# Run the MLE process for each sample size
for n in sample_sizes:
    # Generate random samples
    samples = np.random.normal(loc=true_mean, scale=true_std, size=n)
    
    # Compute the MLE of the mean (sample mean)
    mle_mean = np.mean(samples)
    
    # Store the MLE estimate for this sample size
    mle_estimates.append(mle_mean)

# Plotting the MLE estimates as the sample size increases
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, mle_estimates, label="MLE of Mean", marker='o', linestyle='-')
plt.axhline(y=true_mean, color='r', linestyle='--', label="True Mean")
plt.xlabel("Sample Size (n)")
plt.ylabel("MLE of Mean")
plt.title("Consistency of MLE: Mean Estimate Converging to True Mean")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# True parameters
true_mean = 5
true_std = 2

# Number of samples to simulate the process
max_sample_size = 1000
step_size = 10  # Increase the sample size by this amount in each step

# Arrays to hold sample sizes and their corresponding MLE estimates of the mean
sample_sizes = np.arange(step_size, max_sample_size + 1, step_size)
mle_estimates = []

# Run the MLE process for each sample size
for n in sample_sizes:
    # Generate random samples
    samples = np.random.normal(loc=true_mean, scale=true_std, size=n)
    
    # Compute the MLE of the mean (sample mean)
    mle_mean = np.mean(samples)
    
    # Store the MLE estimate for this sample size
    mle_estimates.append(mle_mean)

# Calculate the rolling average to smooth fluctuations
window_size = 20  # Adjust the window size as needed for more or less smoothing
smoothed_estimates = np.convolve(mle_estimates, np.ones(window_size) / window_size, mode='valid')

# Adjust sample sizes to match the smoothed estimates
smoothed_sample_sizes = sample_sizes[:len(smoothed_estimates)]

# Plotting the MLE estimates and the smoothed version
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, mle_estimates, label="MLE of Mean", marker='o', linestyle='-', alpha=0.5)
plt.plot(smoothed_sample_sizes, smoothed_estimates, label="Smoothed MLE of Mean", color='green', linewidth=2)
plt.axhline(y=true_mean, color='r', linestyle='--', label="True Mean")
plt.xlabel("Sample Size (n)")
plt.ylabel("MLE of Mean")
plt.title("Consistency of MLE: Smoothed Mean Estimate Converging to True Mean")
plt.legend()
plt.show()
```
