```python
import h5py
from IPython.display import display, HTML

# Path to your NWB file
nwb_file_path = "/home/maria/NeuroGarage/000140/sub-Jenkins/sub-Jenkins_ses-small_desc-train_behavior+ecephys.nwb"

def traverse_hierarchy(name, obj, level=0):
    """Helper function to build a nested HTML structure for the file hierarchy."""
    indent = "&nbsp;" * 4 * level
    if isinstance(obj, h5py.Group):
        return f"{indent}<details><summary><b>Group</b>: {name}</summary>\n" + \
               "".join(traverse_hierarchy(k, v, level + 1) for k, v in obj.items()) + \
               "</details>\n"
    elif isinstance(obj, h5py.Dataset):
        return f"{indent}<b>Dataset</b>: {name}<br>\n"
    return ""

# Build the HTML representation
html_structure = "<div style='font-family: monospace;'>\n"
with h5py.File(nwb_file_path, 'r') as f:
    for key, item in f.items():
        html_structure += traverse_hierarchy(key, item)
html_structure += "</div>"

# Display the HTML in the notebook
display(HTML(html_structure))
```
