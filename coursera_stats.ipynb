```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of n values
n_values = np.arange(1, 1001)  # From n=1 to n=1000

# Compute E[X_n^2] for each n
E_Xn_squared = 1 + 1/n_values**2 - 1/n_values**4

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, E_Xn_squared, label=r'$\mathbb{E}[X_n^2]$', color='blue')
plt.axhline(y=1, color='red', linestyle='--', label='Limit = 1')

plt.title(r'Visualization of $\mathbb{E}[X_n^2]$ vs $n$')
plt.xlabel('$n$')
plt.ylabel(r'$\mathbb{E}[X_n^2]$')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
