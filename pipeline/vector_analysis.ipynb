```python
import numpy as np
```

```python
data_dict = {}
for i in range(0,60):
    data_dict[i]=np.load('/home/maria/NeuroGarage/pipeline/weights_to_analyze/weights_'+str(i)+'.npy')
```

```python
first_neuron=np.array([data_dict[i][0] for i in range(0,60)])
print(first_neuron.shape)
```

```python
import matplotlib.pyplot as plt

plt.plot(np.std(first_neuron, axis=0)/np.mean(first_neuron, axis=0))
```

```python
import numpy as np 
import pylab 
import scipy.stats as stats

measurements = np.random.normal(loc = 20, scale = 5, size=100)   
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```
