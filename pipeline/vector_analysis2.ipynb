```python
import numpy as np
data_dict = {}
for i in range(0,60):
    data_dict[i]=np.load('/home/maria/NeuroGarage/pipeline/weights_to_analyze4/weights_'+str(i)+'.npy')
```

```python
import matplotlib.pyplot as plt 
def analyze_random_neurons(data_dict):
    import random
    n_neurons=data_dict[0].shape[0]
    print(n_neurons)
    random_neurons=random.sample(range(0,n_neurons), 20)
    for i in random_neurons:
        print('neuron:',i)
        X=[data_dict[j][i] for j in range(0,60)]
        plt.imshow(np.corrcoef(X))
        plt.colorbar()
        plt.show()
        print(' ')
analyze_random_neurons(data_dict)
```
